```python
"""
DATAPLOT4
"""

import matplotlib.pyplot as plt
import numpy as np

t = [0, 20, 40, 60, 80, 100, 120]
T = [28, 30, 35, 43, 55, 70, 88]
P = [4.8, 8.3, 8.5, 14.5, 37.9, 60.4, 91.6]

fig, ax1 = plt.subplots()      # ---- set up a plot ax1 with subplots

ax1.grid(True)                 # ---- draw grid

ax1.set_xlabel('Time (hrs)')  # ---- label x axis of ax1

line1 = ax1.plot(t, T, 'r', label='Temperature')  # ---- plot temperature in red as curve 1
ax1.set_ylim([0, 100])        # ---- set y axis limits of ax1
ax1.set_ylabel('Temperature (°C)', color='r')     # ---- label y axis of ax1

ax2 = ax1.twinx()             # ---- set up ax2 as twin of ax1

line2 = ax2.plot(t, P, 'b', label='Pressure')     # ---- plot pressure in blue as curve 2
ax2.set_ylim([0, 100])        # ---- set y axis limits of ax2
ax2.set_ylabel('Pressure (psi)', color='b')       # ---- label y axis of ax2

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')       # ---- plot legend

plt.title('Test Data')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Konfigurasi plot
plt.figure(figsize=(10, 6))
plt.axis([0, 140, 0, 100])
plt.axis("on")
plt.grid(True)

# Data input
t = [20, 40, 60, 80, 100, 120]        # Waktu (6 titik)
T = [30, 35, 43, 55, 70, 85]          # Temperatur
P = [2, 3, 4, 5, 3.9, 6]              # Tekanan (6 titik)
V = [16, 58, 54, 46, 35, 2]           # Volume (disesuaikan agar positif)

# Skala data
PP = [p * 10 for p in P]  # Tekanan x10
VV = [v * 1.0 for v in V]  # Volume x1

# Plot data utama
plt.plot(t, T, "r-o", label="Temperature", markersize=8, linewidth=2)
plt.plot(t, PP, "b-s", label="Pressure", markersize=8, linewidth=2)
plt.plot(t, VV, "g-d", label="Volume", markersize=8, linewidth=2)

# Label sumbu utama
plt.xlabel("Time (hrs)", fontsize=12)
plt.ylabel("Temperature (°C)", color="r", fontsize=12)
plt.tick_params(axis="y", labelcolor="r")

# Skala sekunder - TEKANAN (kiri)
for y in np.arange(0, 101, 20):
    plt.text(135, y, f"{y/10}", color="b", va="center", fontsize=10)
plt.text(138, 50, "Pressure (psi)", rotation=90, color="b", fontsize=12)

# Skala sekunder - VOLUME (kanan)
for y in np.arange(0, 101, 20):
    plt.text(140, y, f"{y/100:.1f}", color="g", ha="left", fontsize=10)
plt.text(143, 50, "Volume (cu3)", rotation=90, color="g", fontsize=12)

# Garis bantu volume
for y in np.arange(0, 100, 3):
    plt.plot([137, 139], [y, y], color="g", lw=0.5)

# Legenda dan judul
plt.legend(loc="upper left", fontsize=12)
plt.title("Compression Test Results", fontsize=14, pad=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data input
t = np.array([20, 40, 60, 80, 100, 120])  # Waktu (hrs)
v = np.array([0.6529, 0.6547, 0.7571, 0.7572, 0.7320, 0.7300])  # Volume (cm³)

# ========== PERHITUNGAN REGRESI ==========
n = len(t)
sum_t = np.sum(t)
sum_v = np.sum(v)
sum_tv = np.sum(t * v)
sum_t2 = np.sum(t ** 2)

# Hitung parameter A (slope) dan B (intercept)
A = (n * sum_tv - sum_t * sum_v) / (n * sum_t2 - sum_t ** 2)
B = (sum_v - A * sum_t) / n

# Garis regresi
t_fit = np.linspace(t[0], t[-1], 100)
v_fit = A * t_fit + B

# Hitung RMS error
errors = v - (A * t + B)
rms = np.sqrt(np.mean(errors ** 2))

# ========== VISUALISASI ==========
plt.figure(figsize=(10, 6))
plt.axis([0, 140, 0.64, 0.77])
plt.grid(True, linestyle='--', alpha=0.5)

# Plot data dan garis regresi
plt.scatter(t, v, color='g', s=80, marker='D', label='Data Volume')
plt.plot(t_fit, v_fit, 'g--', linewidth=1.5, label='Regresi Linier')

# ========== ANOTASI ==========
# Persamaan regresi
plt.text(60, 0.68, r'$v = {:.5f}t + {:.5f}$'.format(A, B), color='g', fontsize=12, fontweight='bold')
plt.arrow(78, 0.68, 15, 0.02,
          head_length=5, head_width=0.005,
          color='g', linewidth=1.2)

# Nilai parameter
plt.text(5, 0.66, f'A = {A:.5f}', color='g', fontsize=10,
         bbox=dict(facecolor='white', edgecolor='g', boxstyle='round'))
plt.text(5, 0.645, f'B = {B:.5f}', color='g', fontsize=10,
         bbox=dict(facecolor='white', edgecolor='g', boxstyle='round'))

# RMS error
plt.text(120, 0.65, f'RMS Error = {rms:.4f}', color='g', fontsize=10,
         ha='right', va='center',
         bbox=dict(facecolor='white', edgecolor='g', pad=4))

# Label sumbu
plt.xlabel('Time (hrs)', fontsize=12, fontweight='bold')
plt.ylabel(r'Volume (cm³)', color='g', fontsize=12, fontweight='bold')
plt.tick_params(axis='y', labelcolor='g')

plt.title('Regresi Linier pada Data Volume', fontsize=14, pad=15)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
