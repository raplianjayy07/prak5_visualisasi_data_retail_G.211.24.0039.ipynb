```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75, 75, 50, -50])
plt.axis('on')
plt.grid(True)

# Axis arrows
plt.arrow(0, 0, 20, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, 20, head_length=4, head_width=3, color='k')
plt.text(16, -3, 'x')
plt.text(-5, 17, 'y')

# ================= GREEN circle =================
xc = 0
yc = 0
r = 40

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100
xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)

for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    if p > 90 * np.pi / 180 and p < 270 * np.pi / 180:
        plt.plot([xlast, x], [ylast, y], color='g', linestyle=':')
    else:
        plt.plot([xlast, x], [ylast, y], color='g')
    xlast = x
    ylast = y

plt.scatter(xc, yc, s=15, color='g')

# ================= RED circle =================
xc = -20
yc = -20
r = 10

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100

xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)

for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    plt.plot([xlast, x], [ylast, y], linewidth=4, color='r')
    xlast = x
    ylast = y

plt.scatter(xc, yc, s=15, color='r')

# ================= PURPLE circle =================
xc = 20
yc = 20
r = 50

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100

xlast = xc + r * np.cos(p1)
ylast = yc + r * np.sin(p1)

for p in np.arange(p1, p2 + dp, dp):
    x = xc + r * np.cos(p)
    y = yc + r * np.sin(p)
    plt.plot([xlast, x], [ylast, y], linewidth=2, color=(.8, 0, .8))
    xlast = x
    ylast = y

plt.scatter(xc, yc, color=(.5, 0, .5))

# ================= BLUE DISC (concentric circles) =================
# Tambahkan definisi variabel di sini
xc = -53
yc = -30
r1 = 0
r2 = 10
dr = 1

p1 = 0 * np.pi / 180
p2 = 360 * np.pi / 180
dp = (p2 - p1) / 100

xlast = xc + r1 * np.cos(p1)
ylast = yc + r1 * np.sin(p1)

for r in np.arange(r1, r2, dr):
    for p in np.arange(p1, p2 + dp, dp):
        x = xc + r * np.cos(p)
        y = yc + r * np.sin(p)
        plt.plot([xlast, x], [ylast, y], linewidth=2, color=(0, 0, .8))
        xlast = x
        ylast = y

# Tampilkan hasil plot
plt.show()
```

```python
"""
2DROTRECTANGLE
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([-10, 150, 100, -10])
plt.axis('on')
plt.grid(True)

# __________________________________________axes
plt.arrow(0, 0, 40, 0, head_length=4, head_width=2, color='b')
plt.arrow(0, 0, 0, 40, head_length=4, head_width=2, color='b')
plt.text(30, -3, 'Xg', color='b')
plt.text(-8, 34, 'Yg', color='b')

# __________________________________________center of rotation
xc = 75
yc = 50
plt.plot([xc - 40, xc + 60], [yc, yc], linewidth=1, color='grey')  # --X
plt.plot([xc, xc], [yc - 40, yc + 45], linewidth=1, color='grey')  # --Y
plt.text(127, 48, 'X')
plt.text(70, 90, 'Y')

plt.scatter(xc, yc, s=20, color='k')  # plot center of rotation
plt.text(70, 49, 'c')

# Fungsi rotasi
def rotz(xp, yp, rz):
    c11 = np.cos(rz)
    c12 = -np.sin(rz)
    c21 = np.sin(rz)
    c22 = np.cos(rz)
    xpp = xp * c11 + yp * c12
    ypp = xp * c21 + yp * c22
    xg = xc + xpp
    yg = yc + ypp
    return [xg, yg]

# Koordinat sudut persegi panjang relatif ke pusat rotasi
xp_list = [-20, +20, +20, -20]
yp_list = [-5, -5, +5, +5]

# Label titik sebelum rotasi
plt.text(50, 45, '1')  # Titik 1
plt.text(97, 45, '2')  # Titik 2
plt.text(97, 57, '3')  # Titik 3
plt.text(50, 57, '4')  # Titik 4

# Plot sudut sebelum rotasi
for i in range(4):
    plt.scatter(xc + xp_list[i], yc + yp_list[i], s=10, color='k')

# Gambar persegi panjang awal (belum diputar)
xg = [xc + xp for xp in xp_list] + [xc + xp_list[0]]
yg = [yc + yp for yp in yp_list] + [yc + yp_list[0]]
plt.plot(xg, yg, color='k')

# Rotasi
rz = 45 * np.pi / 180  # konversi ke radian
rotated_points = [rotz(xp_list[i], yp_list[i], rz) for i in range(4)]

# Tambahkan titik pertama di akhir agar tertutup
rotated_points.append(rotated_points[0])

# Pisahkan x dan y
xg_rot, yg_rot = zip(*rotated_points)

# Gambar persegi panjang setelah rotasi
plt.plot(xg_rot, yg_rot, color='r')

# Tampilkan
plt.show()
```
